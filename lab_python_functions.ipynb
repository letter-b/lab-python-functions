```python
#Define a function named initialize_inventory that takes products as a parameter. Inside the function, implement the code for initializing the inventory dictionary using a loop and user input.
def initialize_inventory(products):
    inventory = {}
    for item in products:
        quantity = input(f'Number of {item} available in stock: ')
        while not quantity.isdigit():
            quantity = input(f'Please enter a valid number of {item} available in stock: ')
        inventory[item] = int(quantity)  # Convert the valid quantity to an integer
    return inventory
```

```python
def update_inventory(customer_orders, inventory):
    for item in customer_orders:
        if item in inventory:
            inventory[item] -= 1
            # Ensure the inventory does not drop below zero
            if inventory[item] < 0:
                inventory[item] = 0
            print(f"One {item} removed from inventory.")
        else:
            print(f"There is no {item} in our inventory.")

    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
```
